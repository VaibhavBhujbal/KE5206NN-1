```python
#Import the required libraries
import numpy as np
np.random.seed(1338)

import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
# Load the training and testing data
(X_train, Y_train), (X_test, Y_test) = fashion_mnist.load_data()
# Display purpose:
X_train_orig = X_train
X_test_orig = X_test
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Representative image from each category

```python
image_ids_to_show = []
current_required_label = 0
for idx, label in enumerate(Y_train):
    if label == current_required_label:
        image_ids_to_show.append(idx)
        current_required_label += 1

plt.figure(figsize=(16,8))
num_images = len(image_ids_to_show)
for i in range(num_images):
    plt.subplot(1, num_images, i+1)
    
    plt.imshow(X_train_orig[image_ids_to_show[i]], interpolation='nearest')
    plt.text(0, 0, str(i), color='black', 
             bbox=dict(facecolor='white', alpha=1))
    plt.axis('off')
```
